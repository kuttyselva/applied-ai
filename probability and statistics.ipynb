```python
import numpy as np
import pylab
import scipy.stats as stats
#n(0,1) distribution values
std_nor=np.random.normal(loc=0,scale=1,size=1000)
#print percentiles of std dev
#for i in range(1,101):
   # print(i,np.percentile(std_nor,i))
stats.probplot(std_nor,dist='norm',plot=pylab) 
pylab.grid()
pylab.show()
```

```python
#generate 100 samples
measurement=np.random.normal(loc=20,scale=5,size=100)
#QQ plot
stats.probplot(measurement,dist='norm',plot=pylab)
pylab.grid()
pylab.show()
```

```python
#100 samples from low high
measure=np.random.uniform(low=-1,high=1,size=10000)
stats.probplot(measure,dist='norm',plot=pylab)
pylab.grid()
```
