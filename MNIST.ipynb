d0=pd.read_csv('./train.csv')
print(d0.head(5))
l=d0['label']
d=d0.drop('label',axis=1)
 


```python
print(d.shape)
print(l.shape)
```

```python
#first pick 15 k data points to work on for time efficiency.
#exercise: perform the same on all 42k datapoints
lables=l.head(15000)
data=d.head(15000)
print('shape of sample data =',data.shape)
```

```python
#standardizing the data
#it means on each feature subtract the value by its mean and divide by its varience 
from sklearn.preprocessing import StandardScaler
standardized_data=StandardScaler().fit_transform(data)
print('standardized data shape',standardized_data.shape)
```

```python
#finding covariance matrix A^T *A
sample=standardized_data
#matrix multiplication using numpy
covar=np.matmul(sample.T,sample)
print('co variance matrix shape',covar.shape)
```

```python
#findingg the top 2 eigen values and eigen vectors for projecting into 2d space
from scipy.linalg import eigh
#the parameter 'eigvals' is defined(low to high)
#eigh function will return eigh val and eigh vec  in ascending order
#this code generate top 2 eigh val , vect
value,vector=eigh(covar,eigvals=(782,783))
print('shape of eigen vector', vector.shape )
#converting the eigen vector into (2,d) shape for easyness of further computation
vector=vector.T
print('updated shape of eigen vector', vector.shape )
print('updated shape of eigen values', value.shape )
#here the vector[1] represent the eigen vector corresponding 1st principal
#here the vector[1] represent the eigen vector corresponding 2nd principal
```

```python
#projecting original data on plane
#formed by two principle eigen vectors by vec-vec multiplication
new_coord=np.matmul(vector,sample.T)
print("resultant shape of data ",new_coord.shape)
```

```python
#appemding label to the 2d projected data
new_coord=np.vstack((new_coord,lables)).T
#creating new dara frame for ploting labeled points
df=pd.DataFrame(data=new_coord,columns=('1st principle','2nd principle','lables'))
print(df.head())
```

```python
import seaborn as sn
sn.FacetGrid(df,hue='lables',size=6)\
    .map(plt.scatter,'1st principle','2nd principle')\
    .add_legend()
plt.grid()
plt.show()
```

## PCA using Scikit-Learn

```python
#initializing the pca
from sklearn import decomposition
pca=decomposition.PCA()
```

```python
#configuring the parameters
#the number of components =2
pca.n_components=2
pca_data=pca.fit_transform(sample)
#pca will contain 2d projects of sample data
print('shape of reduced pca',pca_data.shape)
```

```python
#attaching the label for each 2d data point
pca_data=np.vstack((pca_data.T,lables)).T
#create df to plot these data
pca_df=pd.DataFrame(data=pca_data,columns=('1st principle','2nd principle','lables'))
sn.FacetGrid(pca_df,hue='lables',size=6)\
    .map(plt.scatter,'1st principle','2nd principle')\
    .add_legend()
plt.grid()
plt.show()
```

# PCA for dimensionality reduction (not for visualization)

```python
#pca for dimentionality
pca.n_components=784
pca_data=pca.fit_transform(sample)
perc_exp=pca.explained_variance_/ np.sum(pca.explained_variance_)
cum_var=np.cumsum(perc_exp)

#plot the pca spectrum
plt.figure(1,figsize=(6,4))
plt.clf()
plt.plot(cum_var,linewidth=2)
plt.axis('tight')
plt.grid()
plt.xlabel('n components')
plt.ylabel('cum var explained')
plt.show()
```
