```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,4,6,4])
plt.xlabel("numbers")
plt.ylabel("indices")
plt.title('my plot')
plt.show()
```

```python
a=np.array(np.arange(10))
b=a**2
plt.plot(a,b,'ro')
plt.xlabel("numbers")
plt.ylabel("sqr")
plt.title('my plot')
plt.grid()
plt.show()
```

```python
t=np.arange(0.,5.,0.2)
plt.plot(t,t**2,'b--',label='^2',linewidth=2.0)
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.4,'g',label='^2.4')
plt.grid()
plt.legend()
plt.show()
```

```python
a=[1,2,3,4]
b=[1,4,9,16]
c=[1,2,3,4]
d=[2,4,6,8]
lines=plt.plot(a,b,c,d)
plt.setp(lines[0],color='r')
plt.setp(lines[1],color='g')
plt.grid()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b--')
plt.subplot(212)
plt.grid()
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```
