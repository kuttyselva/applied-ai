```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
iris=pd.read_csv('iris.csv')
print(iris.shape)
```

```python
print(iris.columns)
```

```python
#to get no of matches
iris["species"].value_counts()
```

```python
iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
plt.grid()
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris,hue="species",size=7) \
    .map(plt.scatter,"sepal_length","sepal_width") \
    .add_legend();
plt.show()
```

```python
#pair plot plots ncr no of pair graphs
sns.set_style("whitegrid");
sns.pairplot(iris,hue="species",size=3) 
#sns.pairplot(iris,hue="species",size=5,diag_kind="kde") 
plt.show()
```

```python
#histograms
iris_seto=iris.loc[iris["species"]=="setosa"];
iris_vir=iris.loc[iris["species"]=="virginica"];
iris_ver=iris.loc[iris["species"]=="versicolor"];
plt.plot(iris_seto["petal_length"],np.zeros_like(iris_seto['petal_length']),'o')
plt.plot(iris_ver["petal_length"],np.zeros_like(iris_seto['petal_length']),'o')
plt.plot(iris_vir["petal_length"],np.zeros_like(iris_seto['petal_length']),'o')
plt.show()
```

```python
sns.FacetGrid(iris,hue="species",size=5)\
    .map(sns.distplot,"petal_length")\
    .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris,hue="species",size=5)\
    .map(sns.distplot,"petal_width")\
    .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris,hue="species",size=7)\
    .map(sns.distplot,"sepal_length")\
    .add_legend()
```

```python
sns.FacetGrid(iris,hue="species",size=7)\
    .map(sns.distplot,"sepal_width")\
.add_legend()
```

```python
#need for cumulative distribution func
#visually see the percentage of versicolor have petlen <1.6
```

```python
counts,bin_edges=np.histogram(iris_seto['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
#compute cdf
cdf=np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
counts,bin_edges=np.histogram(iris_seto['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
#compute cdf
cdf=np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
#virginica
counts,bin_edges=np.histogram(iris_vir['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
#compute cdf
cdf=np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
#versicol
counts,bin_edges=np.histogram(iris_ver['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
#compute cdf
cdf=np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
#mean variance std dev
print("Mean")
print(np.mean(iris_seto["petal_length"]))
#mean with outline identifier
print(np.std(iris_seto["petal_length"]))
print(np.std(iris_ver["petal_length"]))
print(np.std(iris_vir["petal_length"]))
```

```python
counts,bin_edges=np.histogram(iris_seto['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))

cdf=np.cumsum(pdf)
plt.grid()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
counts,bin_edges=np.histogram(iris_seto['petal_length'],bins=20,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
```

```python
#median , percentile , quantile iqr , mad
from statsmodels import robust
print("median")
print(np.median(iris_seto["petal_length"]))
print("percentile")
print(np.percentile(iris_seto["petal_length"],np.arange(0,100,25)))
print(np.percentile(iris_seto["petal_length"],90))
print("median")
print(robust.mad(iris_seto["petal_length"]))
```
